```python
Automobilesales_df.drop(["PHONE","ADDRESSLINE1","POSTALCODE","CONTACTLASTNAME","PRODUCTCODE"],inplace=True,axis=1)
```

```python
Automobilesales_df.drop(["CONTACTFIRSTNAME","ORDERLINENUMBER","DAYS_SINCE_LASTORDER"],inplace=True,axis=1)
```

```python
Automobilesales_df["DEALSIZE"].value_counts()
```

```python
plt.figure(figsize=(5,5))
plt.hist(x="DEALSIZE",data=Automobilesales_df,edgecolor='black')
Automobilesales_df=Automobilesales_df.sort_values(by=["DEALSIZE"],ascending=False)
plt.show()
```

```python
Automobilesales_df["PRODUCTLINE"].unique()
```

```python
plt.pie(x=Automobilesales_df["PRODUCTLINE"].value_counts(),autopct="%1.0f%%",labels=['Classic Cars','Planes', 'Vintage Cars', 'Motorcycles',
       'Trucks and Buses', 'Ships', 'Trains'],explode=[0,0,0,0,0,0,0],shadow=True)
plt.show()
```

```python
Automobilesales_df["COUNTRY"].unique()
```

```python
Product_sales= Automobilesales_df.groupby('PRODUCTLINE').agg({'SALES':'sum', 'ORDERNUMBER': 'count'})
Product_sales['AVG_SALES'] =Product_sales['SALES'] / Product_sales['ORDERNUMBER']
Product_sales
```

```python
plt.figure(figsize=(10,5))
Product_sales = Product_sales.sort_values(by ="SALES",ascending = False)
Product_sales["SALES"].plot(kind="bar",rot=45,color="skyblue",edgecolor='black')
plt.xlabel=('PRODUCTLINE')
plt.ylabel=("SALES")
plt.show()
```

```python
Automobilesales_df["COUNTRY"].value_counts()
```

```python
colors=plt.cm.plasma(np.linspace(.7, .3, 20))
plt.figure(figsize=(20,5))
sns.barplot(x=Automobilesales_df["COUNTRY"].unique(),y=Automobilesales_df["COUNTRY"].value_counts(),color="orange",edgecolor='black')
plt.legend(["count of countries"])
plt.show()
```

```python
sns.heatmap(Automobilesales_df.corr(numeric_only=True),annot=True)     # positive relationship - coefficient close to 1 
plt.title=('Correlation Matrix')                                       # Negative relationship - coefficient close to -1 
                                                                       # No relationship - coefficient close to 0
```
