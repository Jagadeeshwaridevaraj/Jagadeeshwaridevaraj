### 1. What is the price of diamonds based on their cut?

```python
diamonds_df.dtypes
```

```python
sns.barplot(data=diamonds_df,x="cut",y="price")
```

### 2. which cut has high proportion value in given data?

```python
plt.pie(x=diamonds_df["cut"].value_counts(),autopct="%1.0f%%",labels=['Ideal','Premium','Good','Very Good','Fair'],explode=[0.5,0,0,0,0],shadow=True)
```

### 3. What are the Prices of diamond based on length(x)?

```python
sns.scatterplot(data=diamonds_df,x="x",y="price")
```

## 4. What are the prices of diamonds in given data?

```python
plt.figure(figsize=(10,5))
plt.hist(x="price",data=diamonds_df)
plt.xlabel("price in US dollars")
plt.ylabel("count")
plt.title("Prices of Diamond")
plt.show()
```

### 5. What is the relationship of variable to variable?

```python
sns.heatmap(diamonds_df.corr(numeric_only=True),annot=True)     # positive relationship - coefficient close to 1 
plt.title('Correlation Matrix')                                # Negative relationship - coefficient close to -1 
                                                               # No relationship - coefficient close to 0
```
